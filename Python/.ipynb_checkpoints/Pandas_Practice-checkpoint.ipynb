```python
#merging, joining and concatenating dataframes
import pandas as pd

df1 = pd.read_csv(r'C:\Users\kinnari\Desktop\Data_Analysis\Python\LOTR.csv')
df1
```

```python
df2 = pd.read_csv(r'C:\Users\kinnari\Desktop\Data_Analysis\Python\LOTR 2.csv')
df2
```

```python
df1.merge(df2)
```

```python
df1.merge(df2,how="inner",on=["FellowshipID",'FirstName'],sort=False)
```

```python
df1.merge(df2,how="outer",on=["FellowshipID",'FirstName'])
```

```python
df1.merge(df2,how="left",on=["FellowshipID",'FirstName'])
```

```python
df1.merge(df2,how="right",on=["FellowshipID",'FirstName'])
```

```python
df1.merge(df2,how="cross") #takes value from one dataframe and compares it to another value in the 2nd table
```

```python
df1.join(df2,on='FellowshipID',how='outer',lsuffix='_Left',rsuffix='_Right')
```

```python
df4 = df2.set_index('FellowshipID').join(df2.set_index('FellowshipID'),lsuffix='_Left',rsuffix='_Right')
```

```python
df4
```

```python
#question = Difference between merge and join in pandas
#-merge is more easy to use 
```

```python
pd.concat([df1,df2],join='outer',axis=1)
```
