```python
#cleaning/standardizing phone numbers
#df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','')

#df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])

#df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))

#df["Phone_Number"] = df["Phone_Number"].astype(str).replace(r'\D', ",regex=True).apply(lambda x: f'{x[:3]}-{x[3:6]}-{x[6:10]}' if x else")
#df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])

#df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','')

#df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','')
#above all produced errors for me, below one worked! 

df = df.astype({'Phone_Number': 'string'}) # convert column to string before performing regex
df.Phone_Number = df.Phone_Number.str.replace(r'[^0-9]','',regex=True) # remove everything except numbers
df

#astype is used to change datatype of a series
#regex - A regular expression (also called regex or regexp) is a way to describe a pattern. It is used to locate or validate specific strings or patterns of text in a sentence, document, or any other character input. 
```

```python
#splitting the address

df[["Street Address", "State", "Zip_Code"]] = df["Address"].str.split(',',expand=True)
df
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes','Y')
df["Paying Customer"] = df["Paying Customer"].str.replace('No','N')
df
```

```python
df = df.replace('N/a','')
df = df.replace('NaN','')
df = df.fillna('')
df
```

```python
#filtering down rows of data - we dont want them
'''for x in df.index:
    if bool(df.loc[x,"Do_Not_Contact"] == "Y"):
        df.drop(x,inplace=True)

df'''

df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')

df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N')
df
```
