```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_highest-grossing_science_fiction_films'

path = requests.get(url)
```

```python
soup = BeautifulSoup(path.text,'html')
```

```python
print(soup)
```

```python
table = soup.find_all('table')[0]
```

```python
print(table)
```

```python
movie_titles = table.find_all('th')
```

```python
movie_titles
```

```python
movie_title_table = [title.text.strip() for title in movie_titles]
```

```python
movie_title_table
```

```python
df = pd.DataFrame(columns=movie_title_table)
```

```python
df
```

```python
column_data = table.find_all('tr')
```

```python
for rows in column_data[1:]:
    data_row = rows.find_all('td')
    individual_row_data = [data.text.strip() for data in data_row]
    length = len(df)
    df.loc[length] = individual_row_data
```

```python
df
```
