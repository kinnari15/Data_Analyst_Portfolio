```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_highest-grossing_science_fiction_films'

page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find_all('table')[0]
```

```python
soup.prettify()
```

```python
table = soup.find('table', class_ = 'wikitable plainrowheaders')
```

```python
print(table)
```

```python
print(soup)
```

```python
movie_names = table.find_all('th')
```

```python
movie_names_table = [title.text.strip() for title in movie_names]
```
